```python
#1.19. Создайте новый столбец, в которому будет указана цена приложений в рублях 
#по сегодняшнему обменному курсу валют

def getUsdRate():
    import requests
    import xml.etree.ElementTree as ET

    data = requests.get('http://www.cbr.ru/scripts/XML_daily.asp?numcode=051').content
    tree = ET.ElementTree(ET.fromstring(data))
    rate = float(tree.getroot().findall('.//Valute[CharCode="USD"]/VunitRate')[0].text.replace(',','.'))
    return rate

rate = getUsdRate()
df["PriceRub"] = df["Price"] * getUsdRate()
df
```

```python
#Вопрос 6: Какими характеристиками обладают самые дорогие приложения 
#(самостоятельно решите какие приложения относятся к категории дорогих)?
import seaborn as sns


# sns.distplot(df.Price.dropna()) 


not_free_count = df[df["IsFree"] == False].shape[0]  # всего 3141 платных приложений
prices_cnt = df[df["IsFree"] == False].groupby(["Price"]).size().to_dict()
prices_cnt_list = sorted([(p, v) for p, v in prices_cnt.items()])
i = 0
cnt = 0
while cnt < not_free_count * 0.7:
    cnt += prices_cnt_list[i][1]
    i += 1
border_price = prices_cnt_list[i][0]  # Дорогими будем считать приложения, входящие в 30% самых дорогих (все, не дешевле border_price)
print(border_price)


df["IsExpensive"] = df["Price"] >= border_price

df.groupby(["IsExpensive"]) [["Sup_devices.num", "Ipadsc_urls.num", "Lang.num", "Vpp_lic"]].agg(["mean", "median"])


# Самые дорогие приложения поддерживают меньше девайсов (в среднем 35.2 против 37.6)
# Самые дорогие приложения имеют больше скриншотов (в среднем 4.33 против 3.63). Но медиана в обоих случаях - 5.
# Самые дорогие приложения поддерживают меньше языков в среднем, но распределение шире (медиана для дорогих - 3 против 1 у бесплатных)
```
