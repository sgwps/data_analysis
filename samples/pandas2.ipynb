# Lambda
Ещё называют "анонимная функция". Анонимные функции могут содержать лишь одно выражение, но и выполняются они быстрее. 

Синтаксис:  _lambda arguments: expression_

В качестве arguments передается список аргументов, разделенных запятой, после чего над переданными аргументами выполняется expression. Если присвоить lambda-функцию переменной, то получим поведение как в обычной функции.

Преимущества lambda-выражений мы получаем, используя lambda в связке с другими функциями.

```python
add_qm = lambda a: a + '???'
```

```python
add_qm('what')
```

```python
raise_to_power = lambda x, y: x ** y
```

```python
raise_to_power (2, 3)
```

---------------------------------

```python
#создаём бинарную переменную, принимающую значение 1, если пассажир плыл один, 0 - если не один
titanic['Alone2'] = titanic['FamilySize'].apply(lambda x: 1 if x==0 else 0)
titanic.head()
```

```python
#создаём бинарную переменную, принимающую значение 1, если пассажир был совершеннолетний, 0 - если нет
titanic['Adult'] = titanic['Age'].apply(lambda x: 1 if x>=18 else 0)
titanic.head(20)
```

# Функция
Представляет собой фрагмент кода, который решает определенную задачу. У функции есть имя и список аргументов, которые она принимает.

```python
#создаём функцию
def get_age_gr(x):
    if x < 21:
        return 'young'
    elif x >=21 and x<60:
        return 'adult'
    elif x>=60:
        return 'old'
#не используем else, чтобы Nan не вошли в группу 3
```

-----------------------------------

```python
#создаём категориальную переменную на основе количественной 
titanic['Age_groups'] = titanic['Age'].apply(get_age_gr)
titanic.head(20)
```

```python
titanic['Age_groups'].value_counts()
```

```python
titanic.Age_groups.dtypes #сейчас тип столбца object, текстовая переменная
```

Для работы с категориальными переменными иногда требуется их перекодировать в числовые, с сохранением текстовых меток, привязанных к соответствующим числам. Это можно сделать с помощью метода Categorical.

```python
titanic['Age_groups'] = pd.Categorical(titanic['Age_groups']) #
```

## 6. Ранжирование 

```python
titanic = pd.read_csv('titanic.csv', index_col=0)
titanic['Age_rank_1'] = titanic['Age'].rank()
titanic.sort_values(by='Age').head(15) 
#самый юный посажир получает ранг 1
#в случае связанных рангов рассчитвается средний ранг
```

```python
titanic['Age_rank_2'] = titanic['Age'].rank(method='max') 
titanic.sort_values(by='Age').head(15)
#самый юный посажир получает ранг 1
#в случае связанных рангов рассчитвается максимальный ранг
```

```python
titanic['Age_rank_3'] = titanic['Age'].rank(method='min') 
titanic.sort_values(by='Age').head(15)
#самый юный посажир получает ранг 1
#в случае связанных рангов рассчитвается минимальный ранг
```

```python
titanic['Age_rank_4'] = titanic['Age'].rank(method='min', na_option='bottom') 
titanic.head(19)
#самый юный посажир получает ранг 1
#в случае связанных рангов рассчитвается минимальный ранг
#если возраст не известен, наблюдение оказывается в конце рейтинга
```

```python
titanic['Age_rank_5'] = titanic['Age'].rank(ascending=False) 
titanic.sort_values(by='Age', ascending=False).head(15)
#самый пожилой пассажир получает ранг 1
```
